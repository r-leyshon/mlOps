## Chapter 1: Intro to MLOps

This course begins by segregating discrete tasks into those associated with ML Dev & MLOps:

### Development Tasks

* Training
* Testing
* Labelling
* Validation
* Model selection...

### MLOps Tasks

* Monitoring performance
* Model deployment
* Model maintenance

A strong point was made in the accumulation of technical debt over time. Particularly when a system in production needs careful maintenance in order to diagnose and patch problems quickly.

## Decoupled ML Model Applications

The need to decouple the model from the application in which it is embedded is made. This means that development can happen seperately, allowing for separate maintainance, development and feature decomissioning. The analogy made was that if the ML application was the chassis of a racecar, then the ML model is like the tyres. Changing the model out when new developments result in greater performance should be achievable with a decoupled model. 

Model decomissioning was discussed briefly. Particularly in regulated industries where the audit trail makes **reproducibility** a priority, such as in algorithmic decisioning of insurance offers.

### Model Life Cycle

4. Model monitoring.

5. Model update.

### MLOps Pipeline Summary

MLOps is an extension of DevOps practices. 

DevOps includes the code required to deploy and monitor the applcation, then an automated monitoring routine, MLOps also introduces the data & data monitoring.

Some notes on terms:

* Pipeline - any automated set of tasks
* Artifacts - any output of a pipeline, Eg a table. 

![ML Ops Pipeline ](www/mlops-phasen.jpg)

* Feature store - the database including the extracted features from the raw input data.

* The pipeline combines the data with the code in the training phase. The trained model is then pushed to the **model registry**, specialised for storing and versioning ML models.

* On each build, model metadata will be produced - data about the model important for versioning and poduct lifecycle. This is pushed to a **metadata store**.

* Finally, the **deployment pipeline** takes the model, metadata and application to production.
